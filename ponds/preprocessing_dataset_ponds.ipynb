```python
import pandas as pd
import numpy as np
```

#### Data Clean Up

```python
sheets_names = ['Ponds_PixelesPrototipos6300', 'Creeks_Pixeles4500', 'Others_Pixels7200']
df = pd.read_excel('data/PixelesPrototipos18900.xls', sheetname=sheets_names[2])
df.shape
```

#### Group by id to get the 3x3 matrices

```python
df = df.drop('X', axis=1)
df = df.drop('Y', axis=1)
grouped_by_id = df.groupby(['id'])
```

```python
for i in np.unique(df['id'].values):
    print(grouped_by_id.get_group((i)))
```

#### These are the limits of each sheet where the ids are repeated

```python
sheet_0 = [(0, 899), (900, 3599), (3600, 6299)]
sheet_1 = [(0, 1799), (1800, 4500)]
sheet_2 = [(0, 3599), (3600, 7200)]
```

```python
new_dataset = pd.DataFrame(columns=['matrix', 'target'])
```

```python
for i in np.unique(df['id'].values):
    print("--->{}".format(i))
    for index_values in sheet_2:
        matrix = grouped_by_id.get_group((i))
        first = matrix.loc[index_values[0]: index_values[1]]
        print(first)
        if not first.empty:
            area = first[['R', 'G', 'B']].values.reshape(3, 3, -1)
            new_dataset = new_dataset.append({'matrix': area.ravel(),
                                              'target': int(np.unique(first['clase']))},
                                              ignore_index=True)    
```
